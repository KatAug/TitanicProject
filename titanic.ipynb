Creating Dataframe for Unclean CSV File

```python
df = pd.read_csv("unclean_titanic_passenger_list.csv")
```

```python
index_name=df.index.name
df.index.name = "Passenger #"
df.head()
```

Formatting/Renaming Column Values (Title Casing)

```python
df["sex"] = df["sex"].str.title()
format_df = df.rename(columns={"pclass": "P_class", "survived": "Survived", "name": "Name", "sex": "Sex", "age": "Age","sibsp": "Sibling/Spouse", "parch": "Parent/Children", "ticket": "Ticket", "fare": "Fare", "cabin": "Cabin", "embarked": "Embarked", "boat": "Boat", "body": "Body", "home.dest": "Home_Dest"})
format_df.head()
```

Change all "NaN" to 0's

```python
format_df.fillna(0, inplace=True)
format_df.head()
```
